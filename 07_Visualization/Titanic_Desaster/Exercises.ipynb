### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic=pd.read_csv('train.csv')
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic=titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()
proportions = [males, females]
```

```python
plt.pie(proportions,labels = ['Males', 'Females'],shadow = False,colors = ['blue','red'],explode = (0.15 , 0),startangle = 90,autopct = '%1.1f%%')#autopct自动添加百分比
plt.axis('equal')
plt.title("Sex Proportion")
plt.tight_layout()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex', fit_reg=False)
lm.set(title = 'Fare x Age')
axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)#分组连续变量散点图
```

### Step 8. Create a histogram with the Fare payed

```python
fare_hist=sns.distplot(titanic.Fare)
fare_hist.set(xlabel = 'Value', ylabel = 'Frequency', title = "fare")
sns.despine()# take out the right and upper borders
```
